```python
import pandas as pd
```

```python
linescores = pd.read_csv("./raw_data/reduced/match_line_scores.csv", sep = ",", parse_dates = ['game_date'])
```

```python
linescores
```

```python
seasons = linescores['season_id'].unique()
```

```python
nba_seasons = pd.DataFrame(columns=['season_id', 'date_start', 'date_end'])
```

```python
for season_id in seasons:
    season_df = linescores[linescores['season_id']==season_id]
    
    min_date = season_df['game_date'].min()
    max_date = season_df['game_date'].max()
    
    temp_data = {'season_id': season_id, 'date_start': min_date, 'date_end': max_date}
    
    # Append the dictionary to the list
    nba_seasons = nba_seasons.append(temp_data, ignore_index=True)
```

```python
nba_seasons.to_csv("./raw_data/reduced/nba_seasons.csv", index=False)
```

```python
nba_seasons
```

```python
player_stats = pd.read_csv("./raw_data/reduced/player_statistics.csv", sep = ",", parse_dates = ['date'])
```

```python
def get_season_id(date):
    season_id = nba_seasons[(date >= nba_seasons['date_start']) & (date <= nba_seasons['date_end'])]['season_id']
    return season_id.values[0] if not season_id.empty else None
```

```python
player_stats['season'] = player_stats['date'].apply(get_season_id)
```

```python
player_stats.to_csv("./raw_data/reduced/player_statistics.csv", index=False)
```
